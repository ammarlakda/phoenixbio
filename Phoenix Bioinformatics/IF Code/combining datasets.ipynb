```python
#combining raw datasets 

import os 
import pandas as pd
import numpy as np
```

```python
#showing file list

dirs = os.listdir("cleanedDatasets")

print(dirs)
```

```python

def combineDatasets(dirs):
    #make another for loop that changes inital dataset that others are appended to 
    for i in range(0, len(dirs)):
        print('Dataset of Datasets iteration:', i)
        #creataes new file name from list above
        filename = "cleanedDatasets/" + dirs[i]

        #first dataset, added label of 1
        print("Labelling first file......:", i)
        df =  pd.read_csv(filename, index_col = 0)
        df['Label'] = 1

        
        
        #Appending rest of raw datasets to first dataset used
        for j in range(0, len(dirs)):
            print('Appending file......:', j)
            offset = dirs[(j + i) % 10]   #Used to wrap around array to append unchecked files
            
            if(offset != dirs[i]):
                df2 = pd.read_csv("cleanedDatasets/" + offset)
                df = pd.concat([df, df2], ignore_index = True)
                print('Successful!')
        
        #Making sure that rest of dataset has label = 0 
        df['Label'] = df['Label'].fillna(0)

        df.drop(['Unnamed: 0'], inplace = True, axis = 1)

        new_filename = "FinalDatasets/" + dirs[i]

    
        #putting back into csv 
        df.to_csv(new_filename)
```

```python
combineDatasets(dirs)
```
